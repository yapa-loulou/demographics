```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
max_children = 8
assert max_children > 1
```

```python
from matplotlib import cm
from matplotlib.colors import ListedColormap
a = cm.get_cmap('tab10', 10)
col_w = a(range(max_children))
colors = {'women': ListedColormap(col_w), 'children': ListedColormap(col_w[1:, :])}
```

```python
date_0 = 1980
date_f = 2020
POP_SIZE_0 = 60*(10**6)
trend_pop = 0.005
pop_size = {date_0: POP_SIZE_0}
for y in range(date_0, date_f):
    pop_size[y+1] = round((1+trend_pop)*pop_size[y])
pop_size = pd.Series(pop_size)
pop_size.head()
```

```python
from copy import deepcopy as dc
fact = 1.03
sym_fact = 1 - (1/fact)
print(1/fact, sym_fact)
```

```python
repart = {date_0: [0.07, 0.07, 0.1, 0.18, 0.22, 0.22, 0.1, 0.04]}
# print(round(sum(repart[date_0]),3))
assert round(sum(repart[date_0]),3)==1
data_women = pd.DataFrame(columns = list(range(max_children)))
child_wom_0 = pd.DataFrame({c: round(repart[date_0][c]*pop_size[date_0]) for c in range(max_children)}, index=[date_0])
data_women = pd.concat([data_women, child_wom_0], axis=0)
for y in range(date_0, date_f):
    repart[y+1] = dc(repart[y])
    s = sum(repart[y+1][(max_children//2-1):max_children])*sym_fact/(max_children//2)
    for c in range(max_children//2-1, max_children):
        repart[y+1][c]/= fact
    for c in range(max_children//2):
        repart[y+1][c]+=s
    assert round(sum(repart[y+1]),3)==1
repart_women = pd.DataFrame(repart, index=list(range(max_children))).transpose()
repart_women.head()
```

```python
data_women = repart_women.multiply(pop_size, axis=0).astype(int)
data_women.head()
```

```python
existing_children = data_women.dot(np.arange(max_children))
existing_children.head()
```

```python
number_children_per_women = existing_children.divide(pop_size, axis=0)
plt.plot(number_children_per_women)
plt.title(f"Evolution du nombre moyen d'enfants par femme entre {date_0} et {date_f}")
plt.show()
```

```python
def plot_repartition(df, title, kind="bar", colormap=None, figsize=(12, 6), stacked=True, normalize=True, **kwargs):
    yt = np.linspace(0,1,21) if normalize else None
    yl = (0,1) if normalize else None
    df.plot(kind=kind, colormap=colormap, stacked=stacked, figsize=figsize, xticks=df.index, yticks=yt, ylim=yl, rot=90, title=title, **kwargs).legend(loc='upper left', bbox_to_anchor=(0.1, -0.11), ncol=max_children)#legend='reverse',
    plt.show()
plot_repartition(repart_women, colormap=colors['women'], title=f"Nombre d'enfants par femme (en % du nombre total de femmes) entre {date_0} et {date_f}")
plot_repartition(repart_women, colormap=colors['women'], kind='area', title=f"Nombre d'enfants par femme (en % du nombre total de femmes) entre {date_0} et {date_f}")
```

```python
plot_repartition(df=data_women, colormap=colors['women'], title="Rpartition des femmes par nombre d'enfants, en chiffres absolus", normalize=False)
```

```python
data_children = data_women.drop(columns=0).transpose().multiply(np.arange(1, max_children), axis=0).astype(int).transpose()
data_children.head()
```

```python
plot_repartition(df=data_children, colormap=colors['children'], title="Rpartition des enfants par taille de leur fratrie (mme mre), en chiffres absolus", normalize=False)
```

```python
repart_children = data_children.divide(existing_children, axis=0)
repart_children.head()
```

```python
plot_repartition(df=repart_women, kind='area', colormap=colors['women'], title=f"Nombre d'enfants par femme (en % du nombre total de femmes)")
plot_repartition(df=repart_children, kind='area', colormap=colors['children'], title="Rpartition des enfants par taille de leur fratrie (mme mre), en % du total des enfants")
```

```python
plot_repartition(df=repart_women, kind='line', linewidth=3, colormap=colors['women'], figsize=(10,5), normalize=False, title=f"Nombre d'enfants par femme (en % du nombre total de femmes)", stacked=False)
plot_repartition(df=repart_children, kind='line', linewidth=3, colormap=colors['children'], figsize=(10,5), normalize=False, title="Rpartition des enfants par taille de leur fratrie (mme mre), en % du total des enfants", stacked=False)
```
